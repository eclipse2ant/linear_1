# 有限次元線形空間

- さて、これからは断らない限り、線形空間は有限次元とします。有限次元の時、いつでも基底があります。そして有限次元線形空間は次元が$n$ ならあとでわかりますが $K^n$ に同型です。だから非常に線形空間は簡単なのです。だからむしろ問題は線形写像の方になります。圏論で言えば、対象は単純で射の方が問題なわけです。とはいっても、線形写像は、あとでわかりますが行列なのです。さて、ここで、いつも対面授業では、だれか女子を人に任意に選んで魔助役にします。そして魔女の家には魔法がかかっていって、魔法によって家の周りを猫と犬の入った箱を持って家の周りの道を一周する度に、箱にはいっている猫と犬の数が魔法によって増えます。ここでは　ワンとニャンとしておきますと、ニャンが$x$匹　ワンが$y$匹のとき、それぞれ　$nx, ny+x$匹になります。詳しくは、**Wanko_to_nyanko.pdf** を見てください。そしてそこに書かれてる問いをやってみてください。

- 自分解いたうえで、**Wanko_to_nyanko-2.pdf** を見られてください、要は行列を使えば、$y=Ax$ の形にかけて一次変換、もっと平たく言えば比例の式の2次元版です。にゃんことわんこで二つの量になって、ベクトルにそして比例係数は数から行列に変わっただけです。これだけで、それなりに難しくなりましたよね。でも行列とベクトルを使えば、比例というすっきりした形になります。

- 線形写像は要は多次元版比例です。ただの数の時より、それなりに難しくなってますが、多変量を扱う際の比例ですから、多変量でいちばんやさしい場合です。だから多変数の微分は、微分係数として行列が出てくるわけです。つまり行列とは、線形写像とみれて、そうなると、いままでやった抽象論が使えます。少しは概論で行列を計算したわけですが、さらに意味づけして深めていきます。

- 逆行列は求めることは、すでに知ってるわけです。ではAのｎ乗はどうでしょうか。わんにゃん問題はn乗を求める問題です。その際、SとNに分解して計算しました。実は、こういうのの直和で一般の場合も行えます。これは線形IIでやってくれるのかな、いわゆるジョルダンの標準形です。

- 基本はわんにゃん問題でS は　Semisimple part Nは　Nilpotent part 要は体格成分だけの部分、ここは固有値とか固有ベクトルの話、そしてNは冪零（Nilpotent)行列　基本は　わんにゃんで、それの直和、なんかここで抽象論でやった直和がでてきて、それでなーーんとなく分かった気分になる、それでこそ抽象論の御利益（ごりやく）です。ありがたや。
