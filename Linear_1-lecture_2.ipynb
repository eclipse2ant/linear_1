## 線形空間の例と体

- 線形空間の定義を初めて見て、驚かれたかもしれません。線形空間の元ないしは要素がベクトルです。つまりベクトルを集めた集合が線形空間なのですが、高校の数学と大学の数学の大きな違いは、何かを扱う際、まずその全体からなる集合を先に考えて、その元として物事を考えることが多いです。実際、線形空間の定義も集合の言葉や写像を使って定義されてると思います。

- ベクトルを集めたものが線形空間で、それの定義をまず見たのですが、それならばそれよりもっと基本的な`数`の定義を考える際も、数の集合を考えていきます。実際 N,Z,Q,R,C で**自然数**、**整数**、**有理数**、**実数**、**複素数**の集合を表します。これらの集合を集合の言葉でとらえるとしたらどうなるでしょうか。写像や順序関係など数と集合で習った言葉で定義を考えてみてください。

- ネット上や本で　**体**や**環**　について調べてみてください。これらは先ほどの数の集合の四則演算だけを着目しています。こられの構造を**代数的構造**と呼びます。（ここでは　体や環の定義が分かればよいです。これらは可換体は類体論やガロア理論、岩澤理論とかに発展するし、環論は可換環論は代数幾何などのアフィンスキームとかずっと奥の深い話と繋がりますが、もっと詳しいことは代数学の選択授業で学習すると思います。とりあえず定義が分かればよろしいです。

- ここで大事なことは四則演算は写像として捉えられ、足し算に関して引き算は　足し算に関しての**逆元**を足す操作、同様に掛け算に関しての**逆元**をかける操作が割り算だということです。足し算を和と掛け算を積と以後は呼びます。和も場合によっては積と呼びます。この辺はおいおいわかると思います。これらの和や積（あるいは総称して積）は写像としてあらわされることに注意してください。

- いったん数の集合が体や環といった**代数的構造**としてみれるようになると、線形空間の定義も見え方が変わってくると思います。線形空間は　**和**と**スカラー倍**という二つの演算を持っていますね。ここで、**アーベル群**という概念を調てください。まさに和とはアーベル群の**代数的構造**です。さらにスカラー倍は　線形空間の集合Vとなにか体Kの直積集合からVへの写像として定義される写像です。体Kであって環ではないことに注意してください。この違いはおいおい分かってくルと思います。なお環の作用がある場合には、**加群**といいます。これも代数学の選択授業で扱うと思います。
