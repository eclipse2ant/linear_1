# 行列変形、線形写像の準同型定理、次元定理の幾何的解釈

- **Linear_map-2.pdf** に今までの表現行列と基底による座標表示について、まとめてます。ここで、基底を座標と見て、$K^n$ から $K^m$ への表現行列による1次変換としてあらわされます。行列はむしろ、線形写像の座標表示した**見え方** と見るほうがすっきりします。線形数学概論における行列の変形はすべて、基底の取り換えによる座標表示が変化しただけであって、線形写像自体は変化していない。この変わるものと不変なもの使い分けが重要になってきます。

- １つ例を揚げると行列のランク（階数）があります。定義の仕方はいろいろあります。私がすっきりするのは、行列の基本変形で対角成分だけにしてかつ１か０にするやり方、この１の数が行列のランクです、行列の基本変形は基底の取り換えによる表現行列の変化です。対角にある１の数って、どんなやりかたでも同じ数が出るのでしょうか？　ここで階数とはないかというと、もとの行列を線型写像と見たとき、その写像の像の次元です。だから、表示がいろいろ変わっても写像の表示の変化にすぎません。そして1の数はまさに像の次元になります。まあ、ある意味、1の数が一定になる証明を基底の数の一意性（これが次元が定義できる根拠）という前に言った、線形代数のなかでいちばん面倒な証明に帰着させて逃げたといってもいいかもしれませんが。。。

### 行列の基本変形

- 行列の基本変形に関しては、**Kihon_henkei.jpeg** を見てください（代数入門　堀田著　裳華房から）。基底をどう変換してるかもわかると思います。ここで注意しておくと、**Kihon_henkei.jpeg** にあるように、係数は可換環であればよくて、単元とは積に関して逆元を持つ元、体の場合は0でない元、だから行列式が０でないとき、係数が体の場合は逆行列が存在したわけですが、その証明を眺めれば、Rの場合は単元のとき、逆行列を持ちます。ただしここでは使いませんが、基本変形で行列がきれいな形になるのは、ユークリッド整域のときで、この条件はかなり強い条件です。

### 線形写像の準同型定理、次元定理のお絵かき

- 線形写像の準同型定理をお絵かきしてみましょう。$K=\mathbb{R}$ のときに書いてるわけですが、実際はイメージ的には結構役に立ちます。つまり一般でもそのままでは、嘘ですが、感覚的につかんで、あとは式で表してしまえば、その式自体は一般で成り立つことが多いので、まあお絵かき、幾何的解釈といいましょうか。。。 これについても、**Linear_map-2.pdf** に図を書いておきました。

### 連立方程式の幾何的解釈

- さて連立方程式 $Ax=b$ を考えましょう。これは、$f^{-1}({b})$ を求めよということです。$Ker f$ の作る超平面が　この超平面と原点だけで交わる超平面（図では直線になってますが）これが$V$ の直和分解ですね。そして$Ker f$ をこの原点だけで交わる超平面（図では直線）を平行に$Ker f$ をずらすと $f$ の取る値が変わっていきます。$f(x_0)=b$ なる$V$の点$x_0$をとると、$f^{-1}({b})=x_0 + Ker f$ と表されます。連立方程式を解くとは、この$x_0$と$Ker f$ を求めることです。その際使うのは行を操作する基本変形ですね。これに対応する基底の変換を行って表現行列を変えているのです。これは$b$ の見え方も変えますね。　

### 次元について

- こうやって、幾何的に解釈すると、あまり3次元4次元5次元といっても変わりがないように思えたりもします。確かに有限次元線形空間では、あまり違いがないのですが、幾何学一般ではそうはいきません。ドナルドソン不変量やポアンカレ予想について調べてみてください。3次元と4次元、ないしは高次元においての違いが出て来ます。
